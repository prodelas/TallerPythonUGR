# Matplotlib Tutorial: 5. Gráficos 3D 

El kit de representación 3D introducido en matplotlib a partir de su version 1.0 puede dar lugar a algunos gráficos tridimensionales bastante aparentes y bonitos, consultar el correspondiente manual de [mplot3d](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html) pero aquí de momento tan sólo exploraremos unas cuantas opciones.

```python
# Como siempre, empezaremos importando los módulos habituales
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Este es el toolkit de gráficos 3D 
from mpl_toolkits.mplot3d import Axes3D
```

Justo como antes, cuando creábamos ejes 2D y le asignábamos varios gráficos, ahora también podremos crear algunos ejes 3D (de la clase `axes`) y aplicarles varias rutinas de gráficos 3D, como veremos en los siguientes ejemplos.

## Gráficos de dispersión en 3D 

Los gráficos de dispersión en 3D usarán la misma orden y parámetros que el equivalente en 2D, por lo que le deberá resultar ya familiar. Empezaremos creando unos ejes 3D, a los que ahora necesitaremos pasarle el argumento ``projection='3d'``.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x + y    # en este caso usaremos un código de color propio 

ax.scatter(x, y, z, c=c);
```

##  Lineas curvas en 3D

Lo mismo que con el comando ``scatter`` en 2D y 3D, los equivalentes comandos ``plot`` tendrán la misma para sus argumentos. Así pues el dibujo de curvas en 3D resultará trivial:

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, '-b');
```

## Superficies en 3D

Los gráficos de superficies en 3D, necesitarán el conveniente uso de las correspondientes coordenadas x, y, z, y podrán usarse para realizar interesantes y bonitas formas tridimensionales:

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)
x.shape, y.shape, z.shape
```

Hemos visto aquí también una elegante manera de generar los necesarios arrays bidimensionales que nos permitirán realizar la correspondiente gráfica tridimensional, con algunas de las opciones más habituales: `cmap`, `rstride`, `cstride`, `linewidth` (consultar [mplot3d](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html#line-plots) para más detalles).

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0);
```

## Gráficos  de tipo `Wire-frame` 

Los gráficos con redes de alambre (o de tipo `Wireframe`en inglés) dibujan lineas entre puntos cercanos de la correspondiente superficie tridimensional. Pueden ser mostradas usando el método o comando ``plot_wireframe``. A continuación mostraremos por ejemplo una esfera parametrizada convenientemente:

```python
u = np.linspace(0, np.pi, 30)
v = np.linspace(0, 2 * np.pi, 30)

x = np.outer(np.sin(u), np.sin(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.cos(u), np.ones_like(v)) # consultar np.ones_like
# en el manual de NumPy en caso de no intuir lo que significa
#  https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones_like.html


fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z);
```

# Gráficos de campos vectoriales

Para poder representar campos vectoriales en el espacio usaremos el comando `quiver`. A continuación podemos ver un ejemplo:

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Para construir la malla (`grid` en inglés)
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Para indicar los datos de dirección de las flechas (campo vectorial)
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```
